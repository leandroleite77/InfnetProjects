```python
#Criando o dataframe com os dados dos cliente.
df_customers = spark\
  .read\
  .option("inferSchema", "true")\
  .option("header", "true")\
  .csv("/user/leandrocmleite/dataset_ecommerce/olist_customers_dataset.csv")

df_customers.show(5)
```

```python
# Calculando a quantidade total de clientes por Estado.

df_customers.groupBy("customer_state").count().orderBy('count',ascending=False).show()
```

```python
#Criando o dataframe com os dados dos vendedores (sellers).
df_sellers = spark\
  .read\
  .option("inferSchema", "true")\
  .option("header", "true")\
  .csv("/user/leandrocmleite/dataset_ecommerce/olist_sellers_dataset.csv")

df_sellers.show(5)
```

```python
#Criando o dataframe com os dados dos pedidos.
df_pedidos = spark\
  .read\
  .option("inferSchema", "true")\
  .option("header", "true")\
  .csv("/user/leandrocmleite/dataset_ecommerce/olist_orders_dataset.csv")

df_pedidos.show(5)
```

```python
# Quais são os possíveis status atribuídos a um pedido?

df_pedidos.groupBy("order_status").mean().orderBy("order_status").show()
```

```python
# Quantos pedidos foram aprovados no ano de 2016?

df_pedidos.createOrReplaceTempView("view_pedidos")

df_t = spark.sql("""
SELECT COUNT(*) AS pedidos_aprovados_2017
FROM view_pedidos 
WHERE YEAR(order_approved_at)='2017'
""")

df_t.show()
```

```python
# Quais são os cinco maiores pedidos recebidos na plataforma? Identificar o meio de pagamento, valor, cidade e estado do cliente que realizou os respectivos pedidos.


# 1o Criando o dataframe com os dados de pagamento:
df_pagamento = spark\
  .read\
  .option("inferSchema", "true")\
  .option("header", "true")\
  .csv("/user/leandrocmleite/dataset_ecommerce/olist_order_payments_dataset.csv")

# 2o criando as tabelas para consulta SQL:
df_customers.createOrReplaceTempView("view_customer")
df_pagamento.createOrReplaceTempView("view_pagamento")

df_t = spark.sql("""
SELECT view_pedidos.customer_id, view_pagamento.payment_type, view_pagamento.payment_value, view_customer.customer_city, view_customer.customer_state FROM view_pagamento
INNER JOIN view_pedidos ON (view_pagamento.order_id = view_pedidos.order_id)
INNER JOIN view_customer ON (view_customer.customer_id = view_pedidos.customer_id) ORDER BY view_pagamento.payment_value DESC
LIMIT 5;
""")

df_t.show(5, truncate=False)
```

<h2>Parte 2:</h2>
<h5> Utilizar a biblioteca PySpark para treinar um modelo de aprendizado de máquina </h5>

<h5> Descrição do problema:
Desejamos prever a quantidade de parcelas que serão escolhidas pelo cliente quando utlizado o cartão de crédito como meio de pagamento. Para isso, utilizaremos uma base de dados com as seguintes informações:
    <ul>
    <li>Cliente ID -- customer_id </li>
    <li>Estado -- customer_state </li>
    <li>Cidade -- customer_city </li>
    <li>Prefixo do CEP do cliente -- customer_zip_code_prefix</li>
    <li> Tipo pagamaento realizado -- payment_type
    <li>Quantidade de parcelas -- payment_installments</li>
    <li>Valor da compra -- payment_value</li>
    </ul>

</h5>

```python
# Obter o dataset com as informações necessárias referentes aos cadastro dos clientes, pedidos e meios de pagamento utilizado.

df_order_fin = spark.sql("""
SELECT view_pedidos.customer_id, view_customer.customer_state, view_customer.customer_city, view_customer.customer_zip_code_prefix, view_pagamento.payment_type, view_pagamento.payment_installments,view_pagamento.payment_value FROM view_pagamento
INNER JOIN view_pedidos ON (view_pagamento.order_id = view_pedidos.order_id)
INNER JOIN view_customer ON (view_customer.customer_id = view_pedidos.customer_id) ORDER BY view_pagamento.payment_value DESC;
""")

df_order_fin.show(5, truncate=False)
```

```python
df_order_fin.count()
```

<u> Manipular o dataframe anterior para trabalharmos com um novo dataframe que represente somente os pedidos pagos utilizando cartão de crédito. </u> 

```python
# A partit do dataframe anterir vamos selecionar somente os pedidos que utilizaram como meio de pagamento o cartão de crédito.

df_order_credcard = df_order_fin[df_order_fin['payment_type'] =='credit_card']
df_order_credcard.show(5, False)
```

```python
df_order_credcard.count()
```

```python
# Importar as bibliotecas e módulos necessários para trabalharmos com o modelo de regressão linear
import pyspark
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler, OneHotEncoder, StringIndexer
from pyspark.ml.regression import LinearRegression
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
```

<u>Tratando os dados categóricos:</u>

```python
# Transfor atributos categóricos em numéricos.
# Prefixo CEP tb é tratado como categoria.

state_index = StringIndexer(inputCol="customer_state", outputCol="state_Label")
model_state = state_index.fit(df_order_credcard)
df_order_credcard = model_state.transform(df_order_credcard)

city_index = StringIndexer(inputCol="customer_city", outputCol="city_Label")
model_city = city_index.fit(df_order_credcard)
df_order_credcard = model_city.transform(df_order_credcard)

#zip_index = StringIndexer(inputCol="customer_zip_code_prefix", outputCol="zip_Label")
#model_zip = zip_index.fit(df_order_credcard)
#df_order_credcard = model_zip.transform(df_order_credcard)

#customer_index = StringIndexer(inputCol="customer_id", outputCol="customer_Label")
#model_zip = customer_index.fit(df_order_credcard)
#df_order_credcard = model_zip.transform(df_order_credcard)

df_order_credcard.show(5)
```

```python
#encoder = OneHotEncoder(inputCols=["state_Label","city_Label","zip_Label"], outputCols=["state_vec","city_vec","zip_vec"], dropLast=True)
#encoder = OneHotEncoder(inputCols=["state_Label","city_Label","customer_Label"], outputCols=["state_vec","city_vec","customer_vec"], dropLast=True)

encoder = OneHotEncoder(inputCols=["state_Label","city_Label"], outputCols=["state_vec","city_vec"], dropLast=True)
model_encoder = encoder.fit(df_order_credcard)
df_order_credcard = model_encoder.transform(df_order_credcard)
```

```python
#df_order_credcard.select(["payment_value","state_vec","city_vec","zip_vec"]).show(5)
#df_order_credcard.select(["payment_value","state_vec","city_vec","customer_vec"]).show(5)

df_order_credcard.select(["payment_value","state_vec","city_vec"]).show(5)
```

<u>Criar a coluna com as features que serão utilizadas no modelo.</u>

```python
#assembler = VectorAssembler(inputCols=["payment_value", "state_vec", "city_vec", "zip_vec"],outputCol="features")
#assembler = VectorAssembler(inputCols=["payment_value", "state_vec", "city_vec", "customer_vec"],outputCol="features")

assembler = VectorAssembler(inputCols=["payment_value", "state_vec", "city_vec"],outputCol="features")
df_features = assembler.transform(df_order_credcard)
df_features.select(["payment_installments", "features"]).show(5,truncate=False)
```

```python
df_features.dropDuplicates(['payment_installments']).select('payment_installments').orderBy('payment_installments',ascending=True).show(50)
```

<u>Criar um novo dataframe somente com os dados que serão utilizados no modelo de aprendizagem:</u>

```python
df_novo = df_features.select(["payment_installments", "features"]).withColumnRenamed("payment_installments","label")
df_novo.show(5,truncate=False)
```

```python
df_novo.printSchema()
```

<u> Dividindo a base em conjuntos de treino e teste:</u>

```python
(trainingData, testData) = df_novo.randomSplit([0.8,0.2])
print(trainingData.count(), testData.count())
```

```python
model = LinearRegression().fit(trainingData)
df_result = model.transform(testData)
df_result.show(50,False)
```

```python
#print(model.coefficients)
print(model.intercept)
print(model.summary.meanSquaredError)
print(model.summary.rootMeanSquaredError)
print(model.summary.r2)
```
